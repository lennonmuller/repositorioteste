### O plano cartesiano

Representa um espaço bidimensional (em 2 dimensões), formado por duas linhas perpendiculares, uma horizontal e outra vertical, que se intersetam num ponto. A linha horizontal é chamada eixo das abcissas ou __eixo X__, enquanto a vertical é chamada eixo das ordenadas ou simplesmente o __eixo Y__.
Quanto ao ponto de interseção, é conhecido como o ponto de origem O. __O plano está dividido em 4 quadrantes.__


<img src="./eixos.png"  />



### Pontos e coordenadas

O objetivo é descrever a posição dos **pontos** no plano sob a forma de **coordenadas**, que são formadas associando o valor do eixo X (horizontal) com o valor do eixo Y (vertical).


A representação de um ponto é simples: **P(X,Y)** onde X e Y são a distância horizontal (esquerda ou direita) e vertical (para cima ou para baixo) 
respetivamente, usando como referência o ponto de origem (0,0), mesmo no centro do plano.

<img src="./Puntos.png"  />


### Vetores no plano

Finalmente, um vetor no plano refere-se a um segmento orientado, gerado a partir de dois pontos diferentes.

Para efeitos práticos, é simplesmente uma linha formada a partir de um ponto inicial na direção de outro ponto final, pelo que se entende que um vetor tem comprimento e direção/sentido.

<img src="./Vector.png"  />


Nesta figura, podemos ver dois pontos A e B que poderíamos definir como se segue:
* **A(x1, y1)** => **A(2, 3)**
* **B(x2, y2)** => **B(7, 5)**

E o vetor seria representado como a diferença entre as coordenadas do segundo ponto em relação ao primeiro (o segundo menos o primeiro):
* **AB = (x2-x1, y2-y1)** => **(7-2, 5-3)** => **(5,2)** 

O que significa:

5 à direita e 2 para cima.

#### Experiência
* Criar os pontos A(2, 3), B(5,5), C(-3, -1) e D(0,0) e imprimir.

* Consultar a distância entre os pontos 'A e B' e 'B e A'.

* Criar um retângulo utilizando os pontos A e B.
* Consultar a base, altura e área do retângulo.

````python
import math

#Criando classe Ponto
class Ponto:
    def __init__(self, x=0, y=0):
        self.x = x
        self.y = y
        
        
    #Substituindo str
    def __str__(self):
        return f"({self.x},{self.y})"
    
    
    # Determinando o Quadrante
    def quadrante(self):
        if self.x > 0 and self.y > 0:
            return "1º quadrante"
        elif self.x < 0 and self.y > 0:
            return "2º quadrante"
        elif self.x < 0 and self.y < 0:
            return "3º quadrante"
        elif self.x > 0 and self.y < 0:
            return "4º quadrante"
        else:
            return "Origem"

        
    #Calculo do vetor
    def vetor(self, outro_ponto):
        vetor_x = outro_ponto.x - self.x
        vetor_y = outro_ponto.y - self.y
        return Ponto(vetor_x, vetor_y)

    # Calculo da distancia
    def distancia(self, outro_ponto):
        dx = outro_ponto.x - self.x
        dy = outro_ponto.y - self.y
        return math.sqrt(dx**2 + dy**2)

#Criando classe Retangulo
class Retangulo:
    def __init__(self, ponto1=Ponto(), ponto2=Ponto()):
        self.ponto1 = ponto1
        self.ponto2 = ponto2
    
    # Calculo da base
    def base(self):
        return abs(self.ponto2.x - self.ponto1.x)
    
    # Calculo da Altura
    def altura(self):

        return abs(self.ponto2.y - self.ponto1.y)

    # Calculo da area
    def area(self):
        return self.base() * self.altura()

# Exemplo de uso:
ponto_A = Ponto(2, 3)
ponto_B = Ponto(5, 5)
ponto_C = Ponto(-3, -1)
ponto_D = Ponto(0, 0)

print(f"Ponto A: {ponto_A}")
print(f"Ponto B: {ponto_B}")
print(f"Ponto C: {ponto_C}")
print(f"Ponto D: {ponto_D}")

print(f"Quadrante de A: {ponto_A.quadrante()}")
print(f"Quadrante de C: {ponto_C.quadrante()}")
print(f"Quadrante de D: {ponto_D.quadrante()}")

vetor_AB = ponto_A.vetor(ponto_B)
print(f"Vetor AB: {vetor_AB}")

dist_AB = ponto_A.distancia(ponto_B)
dist_BA = ponto_B.distancia(ponto_A)
print(f"Distância entre A e B: {dist_AB:.2f}")
print(f"Distância entre B e A: {dist_BA:.2f}")

dist_origem = [ponto_A.distancia(Ponto()), ponto_B.distancia(Ponto()), ponto_C.distancia(Ponto())]
mais_afastado = max(dist_origem)
indice_mais_afastado = dist_origem.index(mais_afastado)
pontos = ["A", "B", "C"]
print(f"Ponto mais afastado da origem: {pontos[indice_mais_afastado]}")

retangulo_AB = Retangulo(ponto_A, ponto_B)
print(f"Base do retângulo AB: {retangulo_AB.base()}")
print(f"Altura do retângulo AB: {retangulo_AB.altura()}")
print(f"Área do retângulo AB: {retangulo_AB.area()}")
````

O método deve percorrer a lista, procurar o nome do tipo da classe de cada objecto e encontrar os iguais ao argumento nome, e guardá-las numa nova lista. Esta nova lista deve ser devolvida ao programa principal onde será atravessada e o seu conteúdo exibido. A chamada no programa principal deve ter este aspeto: **``listaCarros = filtrar(lista, “Carro”)``**

* Cria alguns veículos (mais do que um por tipo)
* Filtrar por diferentes tipos de veículos e mostrar os resultados devolvidos.

````python
class Veiculo:
    def __init__(self, cor, rodas):
        self.cor = cor
        self.rodas = rodas

    def __str__(self):
        return f"Veículo (Cor: {self.cor}, Rodas: {self.rodas})"

class Carro(Veiculo):
    def __init__(self, cor, rodas, velocidade, cilindrada):
        super().__init__(cor, rodas)
        self.velocidade = velocidade
        self.cilindrada = cilindrada

    def __str__(self):
        return f"Carro (Cor: {self.cor}, Rodas: {self.rodas}, Velocidade: {self.velocidade} km/h, Cilindrada: {self.cilindrada} cc)"

# Exemplos de carros
carro1 = ("Vermelho", 4, 300, 1496)
carro2 = ("Preto", 4, 150, 225)

print("Descrição do primeiro carro:")
print(carro1)

print("\nDescrição do segundo carro:")
print(carro2)
    

class Moto(Veiculo):
    def __init__(self, cor, rodas, cilindrada):
        super().__init__(cor, rodas)
        self.cilindrada = cilindrada

    def __str__(self):
        return f"Moto (Cor: {self.cor}, Rodas: {self.rodas}, Cilindrada: {self.cilindrada} cc)"

class Caminhao(Veiculo):
    def __init__(self, cor, rodas, carga):
        super().__init__(cor, rodas)
        self.carga = carga

    def __str__(self):
        return f"Caminhão (Cor: {self.cor}, Rodas: {self.rodas}, Carga: {self.carga} toneladas)"

# Criando uma lista de veículos
lista_veiculos = [
    Carro("Vermelho", 4, 300, 1496),
    Moto("Azul", 2, 250),
    Caminhao("Verde", 6, 10)
]

# Função para catalogar os veículos
def catalogar(veiculos, num_rodas=None):
    if num_rodas is not None:
        print(f"\nEncontrados {len(veiculos)} veículos com {num_rodas} rodas:")
    else:
        print(f"\nEncontrados {len(veiculos)} veículos:")

    for veiculo in veiculos:
        if num_rodas is None or veiculo.rodas == num_rodas:
            print(veiculo)

# Testando com diferentes números de rodas
catalogar(lista_veiculos)
catalogar(lista_veiculos, 2)
catalogar(lista_veiculos, 4)



# Função para filtrar os veículos por tipo
def filtrar(veiculos, nome_tipo):
    return [veiculo for veiculo in veiculos if type(veiculo).__name__ == nome_tipo]

# Criando alguns veículos
carro1 = Carro("Vermelho", 4, 300, 1496)
carro2 = Carro("Preto", 4, 150, 225)
moto1 = Veiculo("Azul", 2)
caminhao1 = Veiculo("Verde", 6)

# Criando uma lista de veículos
lista_veiculos = [carro1, carro2, moto1, caminhao1]

# Filtrando por tipo de veículo
lista_carros = filtrar(lista_veiculos, "Carro")
lista_motos = filtrar(lista_veiculos, "Veiculo")
lista_caminhoes = filtrar(lista_veiculos, "Veiculo")

# Exibindo os resultados
print("Carros:")
for carro in lista_carros:
    print(carro)

print("\nMotos:")
for moto in lista_motos:
    print(moto)

print("\nCaminhões:")
for caminhao in lista_caminhoes:
    print(caminhao)
````
